```python
#You must include a written description of three observable trends based on the data.
#1. More drivers per Urban city compared to Suburban and Rural cities.
#2. Even though the average fare is higher in Rural cities, which have less rides per city, 
## Urban cities make more in total fare because they give more frequent rides.
#3. Consistently, Urban cities have the most drivers, the most total fare and the most total rides given.

%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load & Read
city_data = pd.read_csv("city_data.csv")
ride_data = pd.read_csv("ride_data.csv")
#ride_data
#city_data
# Combine the data into a single dataset #on=["city","city"]
city_ride_data_combined = pd.merge(ride_data,city_data, how="left",on="city")
city_ride_data_combined.head()
#city_ride_data_combined_df = pd.DataFrame(city_ride_data_combined)
# Display the data table for preview
#city_ride_data_combined_df.head()
#city_ride_data_combined["city"].value_counts()
#city_data[city_data["city"] == "Richardfort"]
#city_ride_data_combined[city_ride_data_combined["city"] == "Richardfort"]
```

## Bubble Plot of Ride Sharing Data

```python
#Obtain x & y coordinates
## X axis and data, which are the 2 parameters needed to create a scatter plot
## X axis is total number of rides per city & y axis is average fare ($)
## 3 type: Urban, Suburban, Rural
x_group1 = city_ride_data_combined[city_ride_data_combined["type"] == "Urban"] 
x_group1_city = x_group1.groupby("city").ride_id.nunique()
xy_group1_city = x_group1.groupby("city").fare.mean()

x_group2 = city_ride_data_combined[city_ride_data_combined["type"] == "Suburban"] 
x_group2_city = x_group2.groupby("city").ride_id.nunique()
xy_group2_city = x_group2.groupby("city").fare.mean()

x_group3 = city_ride_data_combined[city_ride_data_combined["type"] == "Rural"] 
x_group3_city = x_group3.groupby("city").ride_id.nunique()
xy_group3_city = x_group3.groupby("city").fare.mean()

#Bubble size deteremined by count of drivers per city per type
bubble_drive = city_ride_data_combined.groupby(["type","city"]).driver_count.mean()
bubble_drive

#plt.legend(handles=[urban_plt, suburban_plt], loc="best")
#fig, ax = plt.subplots()
#ax.legend(handles=legend_elements, loc="center")

## Plotting 3 different charts in one scatter
urban_plot, = plt.scatter(x_group1_city,xy_group1_city,c="palevioletred",s=bubble_drive*15, edgecolors="black", label='Urban',alpha=0.75),
suburban_plot, = plt.scatter(x_group2_city,xy_group2_city,c="skyblue",s=bubble_drive*15, edgecolors="black", label='Suburban',alpha=0.75),
rural_plot, = plt.scatter(x_group3_city,xy_group3_city,c="gold",s=bubble_drive*15, edgecolors="black", label='Rural', alpha=0.75),
plt.legend(loc="upper right",markerscale=0.4)
           
plt.grid(color='grey', linestyle='-', linewidth=0.5)
#plt.legend(handles=[legend_elements])


#Titles of chart & axes
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

#Properties of chart
#1. transparency of bubbles 2*. overlay grey grid lines 3*. Upper right hand legend 4*. black outline bubbles 

#5. annotation note
plt.annotate("Note:\nCircle size correlates with \ndriver count per city.",
           xy=(1.2, 0), xytext=(-45, 130),
           xycoords=('axes fraction', 'figure fraction'),
           textcoords='offset points',
           size=10, ha='left', va='bottom')
plt.savefig("Pyber_Ride_Sharing_Data_Scatter_Plot.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#groupby
#count for sizes?
#% of total fare by city type
group_city_data = city_ride_data_combined.groupby("type").fare.sum()
group_city_data
sum_city = (group_city_data/city_ride_data_combined.fare.sum())*100
sum_city
#convert count to percentage?

labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0, 0, 0.1]
# Build Pie Chart
plt.pie(sum_city, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
#plt.axis("equal")
# Save Figure
plt.savefig("total_fares_city_pie_chart.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
group_city_data = city_ride_data_combined.groupby("type").ride_id.sum()
group_city_data
sum_city = (group_city_data/city_ride_data_combined.ride_id.sum())*100
sum_city
#convert count to percentage?

labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0, 0, 0.1]
# Build Pie Chart
plt.pie(sum_city, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")
#plt.axis("equal")
# Save Figure
plt.savefig("total_rides_bycity_pie_chart.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
#Important to use city_data original unmerged data set becuase otherwise count will repeat, as in merged data it 
## duplicates to populate repetition of city (multiple rides per city)
#group_city_data = city_ride_data_combined.groupby("type").driver_count.sum()
#group_city_data
group_city_data2 = city_data.groupby("type").driver_count.sum()
group_city_data
sum_city = (group_city_data2/city_data.driver_count.sum())*100
sum_city
#convert count to percentage?

labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0, 0, 0.1]
# Build Pie Chart
plt.pie(sum_city, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")
#plt.axis("equal")
# Save Figure
plt.savefig("total_drivers_by_city_pie_chart.png")
plt.show()
```
